```python
import numpy as np
import pandas as pd 
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_absolute_error,roc_auc_score,confusion_matrix,classification_report
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import statsmodels.formula.api as sm
import statsmodels.api as smf
```

```python

data = pd.read_csv("alzheimer.csv")
data = data.dropna()
data = data.drop(columns=['DoctorInCharge','PatientID'])
```

```python

data['EducationLevel'] = data['EducationLevel'].astype('category')
data['Ethnicity'] = data['Ethnicity'].astype('category')
```

```python
train_df, test_df = train_test_split(data, train_size=1500,shuffle=False)
```

```python
X_train = train_df.drop(columns=['Diagnosis'])
y_train = train_df['Diagnosis']
X_test = test_df.drop(columns=['Diagnosis'])
y_test = test_df['Diagnosis']
X_train = pd.get_dummies(X_train, drop_first=True)
X_test = pd.get_dummies(X_test, drop_first=True)
```

```python
logit_model = LogisticRegression(max_iter=1000)
logit_model = logit_model.fit(X_train, y_train)
y_pred = logit_model.predict(X_test)
```

```python

X_test = smf.add_constant(X_test)
y_pred_prob  = model.predict(X_test.astype(int))
```

```python
cutoff = y_pred_prob.mean()
cutoff
```

```python
y_pred = np.where(y_pred_prob >= cutoff, 1, 0)
```

```python
conf = confusion_matrix(y_test, y_pred)
print(conf)
```

```python
error_rate = 1 - np.diag(conf).sum() / conf.sum()
print(error_rate)
```

```python
model2 = KNeighborsClassifier(n_neighbors=2)
model5 = KNeighborsClassifier(n_neighbors=5)
model9 = KNeighborsClassifier(n_neighbors=9)

model2.fit(X_train, y_train)
model5.fit(X_train, y_train)
model9.fit(X_train, y_train)

y_pred2 = model2.predict(X_test)
y_pred5 = model5.predict(X_test)
y_pred9 = model9.predict(X_test)

conf2 = confusion_matrix(y_test, y_pred2)
print("K=2")
print('Confusion Matrix\n',conf2)
error_rate2 = 1 - np.diag(conf2).sum() / conf2.sum()
print("Error_rate:",error_rate2)
conf5 = confusion_matrix(y_test, y_pred5)
print("K=5")
print('Confusion Matrix\n',conf5)
error_rate5 = 1 - np.diag(conf5).sum() / conf5.sum()
print("Error_rate:",error_rate5)
conf9 = confusion_matrix(y_test, y_pred9)
print("K=9")
print('Confusion Matrix\n',conf9)
error_rate9 = 1 - np.diag(conf9).sum() / conf9.sum()
print("Error_rate:",error_rate9)
```
